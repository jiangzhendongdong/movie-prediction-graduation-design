```python
import json
import pandas as pd

# 加载 JSON 数据
with open("D:/DoubanMovieRatePredict-main/movie_data.json", "r",encoding="utf-8") as file:
    data = json.load(file)
```

```python
    

# 转换为 DataFrame
df = pd.DataFrame(data)

# 去除双引号
df = df.replace(to_replace=r'\'', value='', regex=True)

df
```

```python

# 保留指定的键
# columns = ["directors", "rate", "cover_x", "star", "title", "url", "casts", "cover", "id", "cover_y"]
columns = ['directors', 'rate', 'cover_x', 'star', 'title', 'url', 'casts', 'cover', 'id', 'cover_y']
df = df[columns]


# 保存为 CSV 文件
csv_file_path = 'D:/DoubanMovieRatePredict-main/movie_data.csv'
df.to_csv(csv_file_path, index=False,encoding="gbk")
csv_file_path
```
